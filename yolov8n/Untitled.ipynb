```python
# plot sample of dataset
import torch
import os

folder = "../../dataset/training_data/images/train"
len(os.listdir(folder))
```

```python
num_3soc = 0
num_other = 0
background_count = 0
for file in os.listdir(folder):
    if file.startswith("co3soc"):
        num_3soc += 1
    elif file.startswith("backg") or file.startswith("000000") :
        background_count += 1   
    else:
        num_other += 1
print(num_3soc)
print(num_other)
print(background_count)

# get a list of only names of files
import os
filename_list = []  
for file in os.listdir(folder):
    #split the file name and the extension into two variables
    filename, file_extension = os.path.splitext(file)
    if filename not in filename_list:
        filename_list.append(filename)
```

```python
lable_folder = "../../dataset/training_data/labels/train"
len(os.listdir(lable_folder))
# remove if no exist image in train folder
for file in os.listdir(lable_folder):
    filename, file_extension = os.path.splitext(file)
    if filename[0]=='.':continue
    if filename not in filename_list:
        os.remove(lable_folder + "/" + file)
        
```

```python
val_label_folder = "../../dataset/training_data/labels/val"
val_image_folder = "../../dataset/training_data/images/val"
val_image_list = []
import os
for file in os.listdir(val_image_folder):
    filename, file_extension = os.path.splitext(file)
    if filename not in val_image_list:
        val_image_list.append(filename)
# remove if no exist image in val image 
removed_count = 0
for file in os.listdir(val_label_folder):
    filename, file_extension = os.path.splitext(file)
    if filename[0]=='.':continue
    if filename not in val_image_list:
        os.remove(val_label_folder + "/" + file)
        removed_count += 1
print(removed_count)
len(os.listdir(val_label_folder))
```
